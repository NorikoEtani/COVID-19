**日本のCOVID-19感染予測：scikit-learn で回帰モデルの結果評価**

ラグの次数を評価法「赤池情報量基準(aic)」あるいは「ベイズ情報量基準(bic)」から選択するための実験

・平均絶対誤差 (MAE:Mean Absolute Error): 実際の値と予測値の絶対値を平均したものであり，MAE が小さいほど誤差が少なく、予測モデルが正確に予測できていることを示す．

・平均二乗誤差 (MSE:Mean Squared Error): 実際の値と予測値の絶対値の2乗を平均したものであり，値が大きいほど誤差の多いモデルといえる．

・二乗平均平方根誤差 (RMSE:Root Mean Squared Error): 観測値と予測値がどの程度離れているのかを評価したものである．

・決定係数 (R2:R-squared, coefficient of determination): モデルの当てはまりの良さを示す指標で，最も当てはまりの良い場合1.0 となるが， 当てはまりの悪い場合マイナスとなることもある．

```python
#必要なモジュールの準備
import numpy as np
import pandas as pd
```

```python
#MAE
from sklearn.metrics import mean_absolute_error
#MSE
from sklearn.metrics import mean_squared_error
#RMSE
from sklearn.metrics import mean_squared_error
#R2
from sklearn.metrics import r2_score
```

**1. Data read**

```python
#Specify date as index 
dateparse = lambda dates: pd.datetime.strptime(dates, '%Y-%m-%d')
```

```python
#Each csv file read
model_patient = pd.read_csv("/content/drive/MyDrive/TimeLineData/Japan_newcases_truth.csv", index_col='date', date_parser=dateparse, dtype='float')
aic_patient = pd.read_csv("/content/drive/MyDrive/TimeLineData/Japan_newcases_aic.csv", index_col='date', date_parser=dateparse, dtype='float')
bic_patient = pd.read_csv("/content/drive/MyDrive/TimeLineData/Japan_newcases_bic.csv", index_col='date', date_parser=dateparse, dtype='float')
model_death = pd.read_csv("/content/drive/MyDrive/TimeLineData/Japan_newdeaths_truth.csv", index_col='date', date_parser=dateparse, dtype='float')
aic_death = pd.read_csv("/content/drive/MyDrive/TimeLineData/Japan_newdeaths_aic.csv", index_col='date', date_parser=dateparse, dtype='float')
bic_death = pd.read_csv("/content/drive/MyDrive/TimeLineData/Japan_newdeaths_bic.csv", index_col='date', date_parser=dateparse, dtype='float')
```

```python
#patient
ts1 = model_patient['new_cases']
ts11 = aic_patient['aic']
ts12 = bic_patient['bic']
#death
ts2 = model_death['new_deaths']
ts21 = aic_death['aic']
ts22 = bic_death['bic']
```

**2. Data evaluation**
